```python
#HAND CRICKET PROGRAM
import random #Imported random Library
print("\nHAND-CRICKET PYTHON PROGRAM WRITTEN BY KUSHAGRA ANIT SINGH\n\nINSTRUCTIONS:\n1. Choose ODD or EVEN \n2. Enter your input for Toss between 1 to 6 \n3. If you win the toss, Opt for Batting or Balling \n4. If Number matches with that of COMP, then it's OUT\n")

def playagain(): #Playagain
    print("-------- \n1. Enter \"Yes\" to Play again. \n2. Enter \"No\" to Exit. \n-------- ")
    PLAYAGAIN = input('Play Again?: ')
    if PLAYAGAIN.upper() == 'YES':
        main()
    elif PLAYAGAIN.upper() == 'NO':
        print("-------- \nThanks for playing!! \nSee you soon... \n--------")
        exit()
    else:
        print('------Invalid input------\n-----Please try again-----')

        playagain()


def main():#Start of the game

    def firstbatting(): #If user or COMP won Toss and opted to bat first
        print("Start batting!..")
        score = 0
        while (True):

            COMP = random.randint(1,6)
            MINE = int(input("Bat: ")) #User Batting Runs Input
            if MINE > 6:
                print('----- INVALID INPUT ------ \n----- GAME RESTARTING -----')
                toss()
            print("COMP Bowls: ", COMP) #COMP Bowling Input
            score = score + MINE #Score Count

            if (COMP == MINE): #Out
                print("Oops...you lost your wicket...!")
                print("Score: ", score)
                break

        print('COMP require', score + 1,'runs to win..!!') #Target
        print('COMP is ready to bat...!') 
        comp_score = 0
        while (True):

            COMP = random.randint(1,6) 
            MINE = int(input("Bowl: "))#User Bowling input
            if MINE > 6:
                print('----- INVALID INPUT ------ \n----- GAME RESTARTING -----')
                toss()
            print("COMP Bats: ", COMP) #COMP Batting Input
            comp_score = comp_score + COMP
            if comp_score > score:
                print("COMP Score", comp_score)
                print('Oops...COMP Won the game!!!') 
                break
            if (COMP == MINE):
                print("WOAH!...COMP GOT OUT!!!")
                print("COMP Score: ", comp_score)
                if score > comp_score:
                    print('Fabulous Victory...Won by',((score)-(comp_score)),'runs, Congratulations buddy...')
                else:
                    print('----- MATCH TIED -----')
                break

    def secondbatting():  #If user or COMP won Toss and opted to bowl first
        print('COMP is ready to start the innings...!')
        comp_score = 0
        while (True):

            COMP = random.randint(1, 6)
            MINE = int(input("Bowl: ")) #User Bowling Input
            if MINE > 6:
                print('----- INVALID INPUT ------ \n----- GAME RESTARTING -----')
                toss()
            print("COMP Bats: ", COMP) #COMP Batting
            comp_score = comp_score + COMP

            if (COMP == MINE): #Out
                print("WOAH!...COMP GOT OUT!!!")
                print("COMP Score: ", comp_score)
                break

        print('COMP is ready to bowl now...get ready!')
        score = 0
        while (True):

            COMP = random.randint(1, 6)
            MINE = int(input("Bat: ")) #User Batting Input
            if MINE > 6:
                print('----- INVALID INPUT ------ \n----- GAME RESTARTING -----')
                toss()
            print("COMP Bowls: ", COMP)
            score = score + MINE
            if comp_score < score:
                print("Score", score)
                print('YOU WON!...Fabulous Victory. Congrats mate!')
                break
            if (COMP == MINE):
                print("Oops...COMP took your wicket...that's OUT!")
                print("Score", score)
                if comp_score > score:
                    print('Oops...COMP Won the game by',(comp_score - score),'runs...')
                else:
                    print('----- TIE MATCH -----')
                break

    def toss():
        TOSS = input('Choose for Toss...ODD or EVEN:  ')
        My_input = int(input('Please enter any number between 1 to 6: '))
        if My_input > 6:
            print('----- INVALID INPUT ------ \n----- GAME RESTARTING -----')
            toss()
        comp_input = random.randint(1, 6)
        print('Input of COMP: ', comp_input)
        if TOSS.upper() == 'EVEN':
            if ((My_input + comp_input) % 2) == 0: #User Toss even
                print('Great...! You won the Toss...What you wanna do to COMP?')
                YOURTOSS = input('BAT or BALL:')
                if YOURTOSS.upper() == 'BAT':
                    firstbatting()
                elif YOURTOSS.upper() == 'BOWL':
                    secondbatting()
                else:
                    print('----- INVALID INPUT ------ \n----- GAME RESTARTING -----') #Invalid
                    toss()
            else:
                compTOSS = random.choice(['bat', 'bowl'])
                if compTOSS == 'bat':
                    print('COMP opted to Bat first...')
                    secondbatting()
                elif compTOSS == 'bowl':
                    print('COMP opted to Bowl first...')
                    firstbatting()
        elif TOSS.upper() == 'ODD': 
            if ((My_input + comp_input) % 2) == 0: 
                compTOSS = random.choice(['bat', 'bowl'])
                if compTOSS == 'bat':
                    print('COMP opted to Bat first..')
                    secondbatting()
                elif compTOSS == 'bowl':
                    print('COMP opted to Bowl first..')
                    firstbatting()

            else:
                print('Great...! You won the Toss...What you wanna do to COMP?')
                YOURTOSS = input('BAT or BOWL: ')
                if YOURTOSS.upper() == 'BAT':
                    firstbatting()
                elif YOURTOSS.upper() == 'BOWL':
                    secondbatting()
                else:
                    print('----- INVALID INPUT ------ \n----- GAME RESTARTING -----')
                    toss()
        else:
            print('----- INVALID INPUT ------ \n----- GAME RESTARTING -----')
            toss()

    toss()
    playagain()

main()
```
